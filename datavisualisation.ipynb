```python
import matplotlib.pyplot as plt
x  = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [1, 3, 4, 3, 1, 6, 5, 4, 1]
y2 = [2, 4, 6, 4, 2, 4, 3, 4, 2]
plt.plot(x, y1, label="line L")
plt.plot(x, y2, label="line H")
plt.plot()

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Graph")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([2, 7, 1, 9])

plt.bar(x, y, color = "#4CAF50")
plt.show()
```
